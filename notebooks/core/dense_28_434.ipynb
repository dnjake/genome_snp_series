<h3>28_434 Hierarchies</h3>
<div style="width:700px">
<p>
Four different hierarchies are selected by the series 28_434.  The associations
that distinguish those hierarchies include:
</p>
<ul>
<li>117_1685 and 123_1561
<li>117_1685 by itself
<li>117_1685 and 13_1696
<li>123_1561 and 13_1696
</ul>
<p>
These hierarchies have been formed through some combination of remodeling of series
SNP expression, recombination events, and selection through the emergence of new series.
</p>
</div>

<h3>117_1685 and 123_1561</h3>
<div style="width:700px">
<p>
The first plot shows data for chromosome samples that express series 28_434 in
association with both 117_1685 and 123_1561.  The 29 samples that
express this association include 25 that resulted from the emergence
of the series 25_27.  SNP counts show that 25 of the samples express
all of the 117_1685 SNPs.  But, 24 of the samples only express 115 of
the 123_1561 SNPs.
</p>
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_28_434, dm.di_117_1685, dm.di_123_1561], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

<h3>117_1685 and 13_1696</h3>
<div style="width:700px">
<p>
This plot shows data for the 74 chromosome samples that express 28_434 along with
117_1685 and 13_1696.  These samples include 73 that express the series
22_73.  SNP counts show that 69 of the samples only express 106 of the
117_1685 SNPs.  All of the samples express large fragments of the
123_1561 SNPs.  The history that lead to this association surely included
a recombination event between a chromosome that expressed 
117_1685, 28_434, 70_166, and 49_136 with another that expressed 13_1696.
The resulting association of series became heavily overexpressed through
some kind of selective advantage introduced with the series 22_73.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_28_434, dm.di_117_1685, dm.di_13_1696], 
                             [dm.di_123_1561], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```

<h3>Only 123_1561</h3>
<div style="width:700px">
<p>
This plot shows data for the 256 28_434 chromosome samples that express 123_1561 but
not 117_1685.  The samples include 251 that express the 180 SNP series
180_251.  SNP counts show that 160 of the samples express all 123_1561 SNPs.
Another 78 express 122 of them.  All of the chromosomes express significant fragments
of 117_1685.  The history that lead to this association included a recombination
event between a chromosome that expressed 123_1561, 28_434, 9_887, and
5_588 with another that expressed 13_1696.  The resulting association
of series became heavily over expressed as a result of some kind of
selective advantage associated with the series 180_251.  But, the
process that generated the 180 SNP series 180_251 series remains unclear.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_28_434, dm.di_123_1561], [dm.di_117_1685], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```
