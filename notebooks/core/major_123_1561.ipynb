<h3>123_1561 Selectors</h3>
<div style="width:700px">
<p>
The series 123_1561 extends across a 435,000 base region of chromosome 2
that includes the gene zranb3 and about half of the gene r3hdm1.  This range
is contained by 117_1685's and located towards its upper end.
<p>
The processes that selected 123_1561 are similar to those that selected 117_1685.
The three processes that generated 193_843, 67_329, and 209_56 all selected
123_1561, 117_1685 and 62_1265.  Most of the chromosome samples that express
either 193_843 or 67_329 express all 123 of the series 123_1561 SNPs.
None of the 56 chromosome samples expressing 209_56 express all 123 SNPs.  But, most
of them do express 122 of those SNPs.
<p>
Two series, 180_251 and 22_35 select 123_1561 without 117_1685.  Most of the
chromosome samples expressing both series express all 123 of the series 123_1561 SNPs
All of those samples express some significant size fragment of 117_1685.
Both of these hierarchies include the series 13_1696 in association with
123_1561.  The 251 chromosome samples that express 180_251 are a large part of the
434 samples that express the series 28_434.  The 35 samples that express
22_35 do not express 28_434.
<p>
Another 25 chromosome samples express an association of series that includes 28_434,
25_27, 123_1561, and 117_1685.  These samples only express 115 or 116 of the
series 123_1561 SNPs.
</div>
