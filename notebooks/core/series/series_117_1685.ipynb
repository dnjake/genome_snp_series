<h3>Series 117_1685</h3>
<div style="width:700px">
<p>
This notebook documents the series 117_1685 and its 117 SNPs. The 1000 genomes data includes 2680
chromosome 2 samples that express none of these 117 SNPs and 1182 samples that express all of them.
Another 294 express 116 of them.  The repeated selection of this series of 117 SNPs is difficult
to explain by a purely random process.
<p>
There does appear to be a background of relatively random fragments of 117_1685.  But, most of
the samples that express more than 1 of the SNPs and less than 116 of them, express one of several
fragments of 117_1685 that are substantially overexpressed in 1000 genomes populations.  This
notebook explores several of these fragments.
</div>

```python
HTML(dm.series_html(dm.di_117_1685))
```

<h3>All chromosome sample 117_1685 SNPs</h3>
<div style="width:700px">
<p>
This plot shows counts of the number of 117_1685 SNPs expressed by different 1000 genomes
chromosome 2 samples.  The distribution of counts is dominated by chromosome samples that
express none of the 117 SNPs and chromosome samples that express all of them.  Given the
large number of SNPs and multiple processes that have selected 117_1685, it is hard
to explain this distribution simply by chance association.
<p>
In the middle of the distribution, there are several cases where fragments of the 117
SNPs are overexpressed among the 1000 genome samples.  Generally, those fragments
can be associated with genetic processes associated with SNP series overexpressed
among the 1000 genome chromosome samples.
</div>

```python
count_data = an.sa_117_1685.unique_snps_per_allele()
plt_obj = smcp(count_data)
plt = plt_obj.do_plot()
show(plt)
```

```python
count_data
```

<h3>117_1685 SNPs</h3>
<div style="width:700px">
<p>
This plot shows the counts of 117_1685 SNPs for chromosome samples that meet
the criteria for being counted as expressing 117_1685.  The distribution
is heavily weighted towards all 117 SNPs or at least 116 of them.  But,
both the 41 samples that express 111 SNPs and the 70 samples that express
106 SNPs are cases where only a very large fragment of the 117 SNPs is really
expressed.  Both of these cases result from recombination events, one at
the high end of the 117_1685 series and the other at the bottom.
</div>

```python
count_data = an.sa_117_1685.unique_snps_per_allele(sm.am_117_1685)
plt_obj = smcp(count_data)
plt = plt_obj.do_plot()
show(plt)
```

```python
count_data
```
