<h3>Series 180_251</h3>
<div style="width:700px">
<p>
This notebook documents the series 180_251 and its 180 SNPs. The 1000 genomes data includes 4626
chromosome 2 samples that express none of these 180 SNPs and 184 samples that express all 180 of them.
Another 25 samples express 179 of them.  The series extends over 720,000 DNA bases in the same part
of chromosome 2 as the series 117_1685, 123_1561, 62_1265, and 193_843.
<p>
All 251 of the chromosome 2 samples that express 180_251 also express 123_1561 and 28_434.  None of
them express 117_1685, 62_1265, or 193_843.  The chromosome samples that express 180_251 are most
of those that express 123_1561 without 117_1685 and establish 123_1561 as a series distinct from
the other overlapping series that it is often associated with.
<p>
While none of the 251 samples express 117_1685, all of them do express significant fragments of it.
241 of the samples express fragments that include 61, 62, or 63 117_1685 SNPs.  The 251 samples are
close to free of any SNPs from 62_1265 or 193_843.
<p>
The data for 28_434 and 180_251 do provide some evidence for the history of 180_251.  That history
started with a chromosome that expressed 117_1685, 123_1561, 9_887, 5_588, and 28_434.  A 
recombination event with a chromosome that expressed 13_1696 was an initial step.  That step
explains the upper region of the 117_1685 SNPs in the fragments associated with 180_251.  But,
some other history must have generated the distributed pattern of the 117_1685 SNPs in the lower
region of the fragments.  Some unknown extended history must have added the 180 SNPs specific
to 180_251 to this base association of series.  The resulting association of series was
further enhanced by the additional series that appeared within the 180_251 hierarchy.  Those
series include 23_226, 6_88, 31_73, 7_21, 15_17, and 10_17.
</div>
