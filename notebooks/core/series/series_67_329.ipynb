<h3>Series 67_329</h3>
<div style="width:700px">
<p>
This notebook documents the series 67_329 and its 67 SNPs. The 1000 genomes data includes 4627
chromosome 2 samples that express none of these 67 SNPs and 275 samples that express all 67 of them.
Another 32 express 66 of them.
<p>
The series 67_329 selects an hierarchy like that selected by the series 193_843 that includes the
series 117_1685, 123_1561, and 62_1265.  The 329 chromosome samples that express 67_329 include
281 that express all 117 SNPs of series 117_1685, 262 that express all 123 SNPs of series 123_1561,
and 215 that express all 62 SNPs of the series 62_1265.  Those samples also include 260 that express
none of the series 193_843 SNPs and only 1 that expresses more than 12 of them.
</div>
