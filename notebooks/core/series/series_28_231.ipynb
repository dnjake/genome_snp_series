<h3>Series 28_231</h3>
<div style="width:700px">
<p>
This notebook documents the series 28_231 and its 28 SNPs. The 1000 genomes data include
include 4648 chromosome 2 samples that express none of these 28 SNPs and 223 that express all
28 of them.
<p>
The series 28_231 has selected an hierarchy that appears to be associated with the emergence
of the series 67_329.  Most chromosome samples that express 28_231 also express 67_329,
62_1265, 123_1561, and 117_1685.  The specific hierarchy selected by 28_231 covers the upper
part of the gene r3hdm1 and the gene ubxn4.  This region of the studied interval of chromosome 2
is beyond the end of 67_329 and in the same part of the interval as 26_1414 and 10_2206 from
the EUR tree.  The series selected by 28_231 in this specific region include 8_328, 9_378,
5_1460, and 10_2206.  This hierarchy is a very different one from the one associated with
10_2206 in the EUR tree root.
