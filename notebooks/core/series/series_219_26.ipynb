<h3>Series 219_26</h3>
<div style="width:700px">
<p>
This notebook documents the series 219_26 and its 219 SNPs. The 1000 genomes data includes 4956
chromosome 2 samples that express none of these 219 SNPs and 20 samples that express all 219 of them.
The series extends over 702,000 DNA bases in the same part of chromosome 2 as the series 117_1685,
123_1561, 62_1265, and 193_843.  All 26 of the chromosome samples that express 209_56 also express 
123_1561, 9_887, and 22_35.  None of them express 117_1685, 62_1265, or 193_843.
<p>
All 26 samples express all 123 123_1561 SNPs.  All of them also express
significant fragments of 117_1685 and 62_1265.  None of them expresses a significant fragment
of the series 193_843.
<p>
219_26 represents another complex history that has selected 123_1561 along with a very large
number of other SNPs.
</div>
