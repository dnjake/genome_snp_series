<h3>Series 4_163</h3>
<div style="width:700px">
<p>
This notebook documents the series 4_163 and its 4 SNPs. The 1000 genomes data include
include 4826 chromosome 2 samples that express none of these 4 SNPs and 163 that express all
4 of them.  All 163 samples also express the series 7_1868 that is also part of the EUR tree root.
162 of the samples express all 7 7_1868 SNPs.  The other 1 expresses 6 of them.
<p>
The series 4_163 is located at the high end of the studied region of chromosome 2.  It covers
the gene dars.  It is common to observe recombination between series in this range and those
in lower parts of the studied region.  Most of the instances of 4_163 are associated with one
of two lower region hierarchies.  87 samples express 4_163 in association with the hierarchy
selected by 67_329.  61 samples express it in association with the hierarchy selected by
180_251.  The other 15 samples express 4_163 and 7_1868 in association with a wide range
of lower region series.
</div>
