<h3>Series 74_210</h3>
<div style="width:700px">
<p>
This notebook documents the series 74_210 and its 74 SNPs. The 1000 genomes data includes 4735
chromosome 2 samples that express none of these 74 SNPs and 201 samples that express all 74 of them.
The series 74_210 extends over more than 600,000 DNA bases in the same region of chromosome 2 as
the series 117_1685, 123_1561, 62_1265, and 193_843.  All 210 of the samples that express 74_210
also express 117_1685.  None of the samples express 123_1561, 62_1265, or 193_843.  The samples
that express 74_210 are the primary contributors to establishing 117_1685 as a distinct series.
<p>
The samples that express 74_210 include 109 that express all 117 117_1685 SNPs.  Another 89 express
116 of those SNPs.  The other 12 samples express counts of 117_1685 SNPs that range from 112 through
115.  None of the 210 samples express significant fragments of 123_1561, 62_1265, or 193_843.  The
samples are close to free of any SNPs from those series.
<p>
The series 74_210 is modestly overexpressed in European and African populations.  But, its expression
is relatively evenly distributed across all of the 1000 genomes populations.
</div>
