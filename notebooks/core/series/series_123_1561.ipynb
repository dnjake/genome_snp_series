<h3>Series 123_1561</h3>
<div style="width:700px">
<p>
This notebook documents the series 123_1561 and its 123 SNPs. The 1000 genomes data includes 2948
chromosome 2 samples that express none of these 123 SNPs and 1186 samples that express all of them.
Another 282 express 122 of them.  The number of SNPs in overexpressed fragments appears to be 
clustered between 106 and 110 SNPs.  
<p>
This notebook explores series that express some of those fragments.  There appear to be three
main sources for the 123_1561 fragments.  One was the recombination event of series 8_267 in the
South Asian tree and the selection of that recombinant by the appearance of the series 9_39.  Another
was the hierarchy selected by the series 290_16.  The other was a variety of remodeling events as
parts of the 123_1561 SNPs were lost in the hierarchies associated with the series 28_434.
</div>
