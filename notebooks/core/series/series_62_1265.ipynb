<h3>Series 62_1265</h3>
<div style="width:700px">
<p>
This notebook documents the series 62_1265 and its 62 SNPs. The 1000 genomes data includes 3234
chromosome 2 samples that express none of these 62 SNPs and 1096 samples that express all of them.
Another 70 that express 61 of them. 
<p>
There are 331 chromosome samples that express  4 or more SNP fragments of series 62_1265.  This
notebook explores the hierarchies associated with some of those fragments
<p>
The fragement expressed by 85 chromosome samples includes only the first 8 62_1265 SNPs.  It resulted
from a recombination event between the South Asian tree and the East Asian tree that associated
the series 4_815 and 5_684.
<p>
Two other hierarchies include fragments of 62_1265.  One is the hierarchy selected by the series
290_16.  The other was selected by the series 219_26.  The 16 chromosome samples that express
290_16 all express a 40 SNP fragement of 62_1265.  The 26 chromosome samples that express
219_26 express either 48 or 47 SNP fragments of 62_1265.
<p>
A significant number of the remaining 62_1265 fragments are part of 28_434 hierarchies.  A number
of those fragments include small numbers of 62_1265 SNPs.
</div>
