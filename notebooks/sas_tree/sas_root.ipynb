<h3>SAS Tree</h3>
<div style="width:700px">
<p>
The plot below shows the association of series that form the root of the SAS tree.  The four
major series 117_1685, 123_1561, 62_1265, and 193_843 all fill a large part of a
629,000 base region that covers the genes rab3gap1, zranb3, and half of r3hdm1.
All but one of the 843 chromosome 2 samples expressing the series 193_843 express 117_1685.
All 843 express both 123_1561 and 62_1265.  All of the 843 samples that express
the series 193_843 express an hierarchy of series that is part of the SAS tree.
No intermediate series have been identified that offer any record of the processes that 
generated the 193 SNPs included in the series 193_843.  The other three series all
are parts of other hierarchies that have experienced histories of selection independent
of the formation of the SAS tree.
<p>
The association of the series 5_1460, 10_1218, 13_911,  4_1149, 9_1170, 9_1023, and 39_1014 form the
the upper region of the SAS tree root.  Like the other hierarchies, it appears likely that all of these
fragments have an history as part of a single series of SNPs that covered the upper
half of the gene  r3hdm1 and the genes ubxn4, lct, mcm6, and dars.  
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_193_843], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```

<div style="width:700px">
<p>
The plot below shows 29 series that are descendants of the series 193_843.  The requirement
for identifying these subset series was relaxed from a match of 0.9 to one of 0.8.  There has been
a recombination event in the history of samples expressing the series 8_267 that causes 44 of them not to express
193_843.  Those samples include 39 that express the series the series 9_39.  See the 
[sas_sub_8_267 notebook](sas_sub_8_267.ipynb) for more analysis.
<p>
The series 4_815 also does not meet the 0.9 subset match criterion.  The cause is a
recombination event with series that are part of the EAS tree.  In both cases,
fragments of the series associated with the SAS tree root are still present and
provide strong evidence for an history of simple recombination events.
</div>

```python
plt_obj = dm.subset_yes_no([dm.di_193_843], min_match=0.8)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```

<div style="width:700px">
<p>

</div>
