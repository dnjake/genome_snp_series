<h3>SAS Subtree 6_713</h3>
<div style="width:700px">
<p>
Out of the 843 chromosome 2 samples that express the SAS tree hierarchy, 713 express
the series 6_713.  The series 6_713 is a more exact root of its subtree than
the series 193_843 is of the SAS tree because a recombination event involving
the series 8_267 has lost enough of the 193 SNPs in the series 193_843 for the
series no longer to be scored as expressed by the recombinant chromosomes.  A selection
process associated with the emergence of the series 9_39 has generated most of the
instances of chromosome samples that express 6_713 but not 193_843.
<p>
All but 5 of the samples that express 6_713 also express 4_815.  The other
samples that express 4_815 are mostly derived from recombination events that
associated instances of 4_815 with the EAS tree.  Recombination events plus some
kind of selection process have resulted in 90 samples that express 4_815 in
association with the series 5_684 as part of the EAS tree hierarchy.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_6_713], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```
