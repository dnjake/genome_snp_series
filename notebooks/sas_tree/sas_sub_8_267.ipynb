<div style="width:700px">
<h3>Series 8_267</h3>
<p>
The plot below shows the series most strongly associated with series 8_267.
Only 223 of the 267 samples that express 8_267 express the series 193_843
that defines the root of the SAS tree.  Most of the samples
that fail to express 193_843 express the series 9_39.  That series has selected
an hierarchy that is the result of a recombination event where enough SNPs 
of the series 193_843 and 123_1561 were lost for the remaining fragments of 
those series to no longer meet the criteria needed to identify them as expressed.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_8_267], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```

<div style="width:700px">
<p>
The plot below shows a basis plot for the samples that express the series 8_267.

137 chromosome samples that don't express any series in the interval more specific then 8_267.
There are 6 more samples that don't express any of the descendant series but do express
both 8_267 and some more specific series in the higher part of the region as a
result of some recombination event.
