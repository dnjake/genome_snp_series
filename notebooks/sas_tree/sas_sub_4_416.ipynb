<div style="width:700px">
<p>
The plot below shows the hierarchy for the series 6_19.  All 19 of the samples
that express it come from African populations.  Only 3 of the 39 samples that express
its parent, 14_39 are not African.  Those 3 American samples probably are a result
of recent American arrivals from Africa.  It is not clear whether 14_39 and 6_19 resulted
from some back migration to Africa or whether a prior population expressing 4_416 as a basis
series has been lost from Africa.  But the data seem to suggest some kind of selective advantage
for 14_39 and 6_19 in Africa and 4_416 by itself elsewhere.
</div>

```python
di_6_19 = 353266
plt_obj = dm.superset_yes_no([di_6_19], min_match=0.9)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```

<div style="width:700px">
<p>
The next three plots show series associatons for hierarchies selected by the series 6_43.
The samples expressing series 6_43 include 30 South Asian, 12 European, 1 American, no East Asian,
and no African.  The hierarchy selected by the appearance of this series is different from most
of those in the SAS tree because of more variation in the series in the upper part of the region
that are associated with series 6_43.  Just over half of the samples that express 6_43 also
express 81_857 and the fragments that are generally associated with it.  These upper region
series are most common as part of the EAS tree root.  Another 10 samples have lost series
81_857 and express a set of upper region fragments that overlap with those generally associated
with it.  Only 4 of the samples express the series 39_1014 that is most commonly associated
with the SAS tree root.
