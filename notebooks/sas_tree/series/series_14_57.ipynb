<h3>Series 14_57</h3>
<div style="width:700px">
<p>
This notebook documents the series 14_57 and its 14 SNPs.  The 1000 genomes data includes
4941 samples that express none of its 14 SNPs and 55 samples that express all 14 of them.
Another 2 samples express 13 of the SNPs. The 14 SNP series covers more than 710,000 DNA bases.
It is over expressed in South Asian populations.  It is not expressed by any samples from
Africa, America, or Europe.
<p>
14_57 is a child of the series 7_90.  It is part of the SAS tree hierarchy that does not express
6_713 or 4_815.  But, all 57 samples do express 193_843, 62_1265, and 123_1561.  56 of them express
117_1685 and 7_90.  1 sample expresses an hiearchy that resulted from a recombination event where
early SNPs from 117_1685, 62_1265, 193_843, and 7_90 were lost.  None of the 57 samples express
more than 122 of the 123_1561 SNPs.
<p>
The series 14_57 extends into the region of chromsome 2 where the upper part of the gene r3hdm1
and the gene ubxn4 are located.  This part of the studied region is where 26_1414 is located
in the EUR tree and 32_1361 in the EAS tree.  All 14_57 samples express the series 55_180,
7_1303, and 4_1442 along with the upper part of 14_57 in this region.
</div>
