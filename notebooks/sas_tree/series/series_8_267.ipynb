<h3>Series 8_267</h3>
<div style="width:700px">
<p>
This notebook documents the series 8_267 and its 8 SNPs.  The 1000 genomes data includes
4688 samples that express none of its 8 SNPs and 259 samples that express all 8 of them.
Another 8 samples express 7 of the SNPs. The 8 SNP series covers more than 471,000 DNA bases.
It is overexpressed in American, European, and South Asian populations.  It is under expressed
in African populations.
<p>
The series 8_267 is a child of 6_713.  All 267 samples express 6_713.  264 of them express 4_815.
But, only 223 samples express 193_843.  The main reason is the history of a recombination event
shared by 41 of the 8_267 samples.  That recombination event caused the loss of the top 23
193_843 SNPs.  38 of the 8_267 samples have resulted from this recombination event and the selection
process associated with the appearance of the series 9_39.
<p>
The descendant series of 8_267 are 5_16, 5_26, 9_39, and 4_45.  There are 143 samples that express
8_267 but none of its descendants.  Those samples include 66 that express all the standard series
of the SAS tree root including 39_1014 along with 5_976 and 6_820.  Another 39 express the same
hierarchy without 5_976 or 6_820.  The other samples show an history of a variety of recombination
events.  3 show the an history of the event associated with 9_39 where recombination with a chromosome
containing 32_1361 and 81_857 replaced the top SNPs in 193_843.  2 show an history of a recombination
event where 13_1696 replaced the lower part of 193_843.
</div>
