<h3>Series 10_25</h3>
<div style="width:700px">
<p>
This notebook documents the series 10_25 and its 10 SNPs.  The 1000 genomes data includes
4981 samples that express none of its 10 SNPs and 23 samples that express all 10 of them.
Another 2 samples express 9 of the SNPs. The 10 SNP series covers more than 457,000 DNA bases.
It is over expressed in South Asian populations.  It is not expressed by any samples from East
Asian or African populations.
<p>
The 25 samples include 17 that express the same complex association of 19 series across the
studied interval.  In the hierarchy formed by those associations 10_25 is a child of 29_68.
29_68 is a direct child of 193_843 without 6_713 or 4_815.
<p>
The other 8 samples show histories of a variety of genetic events.  4 samples have lost the
middle SNP of the series 5_1460.  2 samples have experienced recombination events in the
very top part of the studied region.  1 sample has lost the top 4 SNPs of 29_68 while retaining
all of the 193_843 SNPs.  The most likely explanation is a recombination event with a chromosome
that expressed the 193_843 hierarchy but not 29_68.
</div>
