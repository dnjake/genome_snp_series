<h3>Series 14_39</h3>
<div style="width:700px">
<p>
This notebook documents the series 10_25 and its 10 SNPs.  The 1000 genomes data includes
4961 samples that express none of its 14 SNPs and 33 samples that express all 14 of them.
Another 6 samples express 13 of the SNPs. The 14 SNP series covers more than 650,000 DNA bases.
It is over expressed in African populations.  It is not expressed by any samples from East
Asian, European, or South Asian populations.
<p>
The series 14_39 is part of the SAS tree hierarchy that is rooted in the series 4_416,
6_713, 4_815, 193_843, 6_946, 62_1265, 123_1561, and 117_1685.  All 39 samples express
all of these series.  However 3 samples have lost 10 SNPs from 123_1561 and 12 SNPs from 193_843.
2 of these samples come from an external African population.  The other 1 comes from an
American population.
</div>
