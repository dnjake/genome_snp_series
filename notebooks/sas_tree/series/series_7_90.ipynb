<h3>Series 7_90</h3>
<div style="width:700px">
<p>
This notebook documents the series 7_90 and its 7 SNPs.  The 1000 genomes data includes
4910 samples that express none of its 7 SNPs and 81 samples that express all 7 of them.
Another 9 samples express 6 of the SNPs. The 7 SNP series covers more than 525,000 DNA bases.
It is overexpressed in South Asian populations.  It is not expressed by any samples from European
populations and by only 1 American sample.
<p>
The series 7_90 is a direct child of the series 193_843.  None of its samples express 6_713 or 4_815.
7_90 is the root of a subtree that includes 14_57, 19_27, and 15_27.  Only 6 samples express 7_90
without any of its descendants.  The 6 samples include 2 that express the full hierarchy of the
SAS tree root including 39_1014.  The other 4 show histories of a variety of recombination events.
</div>
