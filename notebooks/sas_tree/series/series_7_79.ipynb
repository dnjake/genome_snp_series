<h3>Series 7_79</h3>
<div style="width:700px">
<p>
This notebook documents the series 7_79 and its 7 SNPs.  The 1000 genomes data includes
4922 samples that express none of its 7 SNPs and 79 samples that express all 7 of them.
The 10 SNP series covers more than 454,000 DNA bases. It is heavily over expressed in 
American populations.  Its samples include 66 American, 11 East Asian, and 2 European.
<p>
The series 7_79 is a child of the series 4_91 and parent of the series 4_38 and 4_27.
Ony 14 samples express 7_79 without 4_38 or 4_27.  Those samples include 9 that express
39_1014 and all of the standard SAS tree root series.  Another 2 samples express the
EUR tree root series 26_1414, 64_1575, 10_2206, and 7_1868.  Another sample expresses
the pair of exceptional EAS series 95_176 and 51_176.
</div>
