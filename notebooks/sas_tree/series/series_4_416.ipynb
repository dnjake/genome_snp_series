<h3>Series 4_416</h3>
<div style="width:700px">
<p>
This notebook documents the series 4_416 and its 4 SNPs.  The 1000 genomes data includes
4560 samples that express none of its 4 SNPs and 416 samples that express all 4 of them.
The 4 SNP series covers more than 306,000 DNA bases.
It is over expressed in South Asian and under expressed in African and European
populations.
<p>
The series 4_416 is the root of an SAS subtree.  It roots almost half of the SAS tree.
All 416 samples express 6_713, 62_1265, 123_1561, and 117_1685.  415 of them express
193_841 and 6_946.  414 express 4_815.
<p>
The series 4_416 descendant series include 4_91, 7_79, 6_43, 4_49  14_39, 4_38, 9_29, 4_27,
and 6_19.  There are 165 samples that express 4_416 but none of its descendants.  These
samples include 84 that express 39_1014 and all of the common series of the SAS tree root.
The other samples have experienced some history of recombination events.
</div>
