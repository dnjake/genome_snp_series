<h3>Series 19_27</h3>
<div style="width:700px">
<p>
This notebook documents the series 19_27 and its 19 SNPs.  The 1000 genomes data includes
4974 samples that express none of its 19 SNPs and 26 samples that express all 19 of them.
Another 1 sample express 8 of the SNPs. The 19 SNP series covers almost 702,000 DNA bases.
26 of the samples come from African populations. 1 comes from an American population.
<p>
The series 19_27 is a child of the series 7_90.  All 27 samples express 7_90, 193_843,
62_1265, 123_1561, and 117_1685.  This notebook explores the variations in the series
associations expressed along with 19_27.
