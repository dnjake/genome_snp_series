<h3>Series 29_68</h3>
<div style="width:700px">
<p>
This notebook documents the series 29_68 and its 29 SNPs.  The 1000 genomes data includes
4922 samples that express none of its 29 SNPs and 63 samples that express all 29 of them.
Another 2 samples express 28 of the SNPs. The 29 SNP series covers almost 611,000 DNA bases.
It is over expressed in South Asian populations.
<p>
The series 29_68 is a direct child of 193_843.  It is the root of an hierarchy that includes
the series 10_25, 6_19, and 8_17.  There are 11 samples that express 29_68 and none of its
descendant series.  Those 11 samples all express 193_843, 62_1265, 123_1561, and 117_1685.
But, the associated series in the higher parts of the region are not the common ones of the
SAS tree.  7 of the samples express 32_1361.  10 express 10_2206.  But, none express 64_1575
or 81_857.
</div>
