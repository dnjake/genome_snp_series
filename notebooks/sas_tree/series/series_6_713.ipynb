<h3>Series 6_713</h3>
<div style="width:700px">
<p>
This notebook documents the series 6_713 and its 6 SNPs.  The 1000 genomes data includes
4278 samples that express none of its 6 SNPs and 706 samples that express all 6 of them.
Another 7 samples express 5 of the SNPs. The 6 SNP series covers more than 275,000 DNA bases.
It is over expressed in South Asian and American populations.
<p>
Most samples that express an SAS tree hierarchy express some subtree of 6_713.  But only
688 of the 6_713 samples express 193_843.  The reason is an history of recombination for
samples expressing the series 8_267.  That history caused some samples to lose 193_843
and 123_1561.  All 446 samples that express 6_713 but not 8_267 express 117_1685,
123_1561, and 62_1265.  445 of them express 193_843.
<p>
Only 26 samples express 6_713 without expressing any of its descendant series.
21 of them express 39_1014.  Those 21 include 11 that express just the standard
series of the SAS tree root.  Another 8 African samples add the series 9_27.
The remaining samples show a variety of recombination histories.
</div>
