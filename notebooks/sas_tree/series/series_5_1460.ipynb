<h3>Series 5_1460</h3>
<div style="width:700px">
<p>
This notebook documents the series 5_1460 and its 5 SNPs.  The 1000 genomes data includes
3375 samples that express none of its 5 SNPs and 1460 samples that express all 5 of them.
The 5 SNP series covers more than 31,000 DNA bases.  It is modestly overexpressed in 
African and South Asian populations.
<p>
The 5_1460 samples include 731 that are associated with 193_843.  Those 731 samples include
729 that express 10_1218, 691 that express 1149, and 685 that express 13_911.  All 4 series
are located within the gene r3hdm1 in the same part of chromosome 2 as 26_1414 in the EUR
tree and 32_1361 in the EAS tree.
<p>
There are 180 samples that express 5_1460, 39_1014, but not 193_843.  Those samples include
180 that express 9_1170, 179 that express 4_1149, 179 that express 9_1023, 179 that express
13_911, and 169 that express 10_1218.  It appears likely that all of these series have a
common history as a single series that broke down into fragments with some indpendent identity.
<p>
The independent identity of 5_1460 is primarily the result of a selection process associated
with the series 4_612.  The 5_1460 samples include 461 that express 4_612.  Those samples
show a completely different pattern of associations in the region of r3hdm1 from those where
5_1460 is associated with 39_1014.
</div>
