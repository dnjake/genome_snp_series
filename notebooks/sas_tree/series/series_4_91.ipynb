<h3>Series 4_91</h3>
<div style="width:700px">
<p>
This notebook documents the series 4_91 and its 4 SNPs.  The 1000 genomes data includes
4906 samples that express none of its 4 SNPs and 91 samples that express all 4 of them.
The 4 SNP series covers more than 571,000 DNA bases.  It is heavily overexpressed in American
populations and modestly overexpressed in East Asian populations.
<p>
The series 4_91 is the root of a subtree that includes 7_79, 4_38, and 4_27.  It is a child
of 4_416.  All 91 samples express the major series in the SAS tree hierarchy associated
with 4_416.
<p>
There are 15 samples that express 4_91 but not 7_79.  Those samples all come from East Asian
populations.  14 of them express all of the series in the SAS tree root including 39_1014.
1 has experienced a recombination event that lost 39_1014.
</div>
