<h3>Series 5_26</h3>
<div style="width:700px">
<p>
This notebook documents the series 5_26 and its 5 SNPs.  The 1000 genomes data includes
4977 samples that express none of its 5 SNPs and 26 samples that express all 5 of them.
The 5 SNP series covers more than 438,000 DNA bases. It is overexpressed in European
populations.
<p>
The series 5_26 is a child of the series 8_267.  Its 26 samples include 15 that express the
series 39_1014.  Those samples include 13 that express the full SAS tree root hierarchy
associated with the series 8_267.  The other two show an history of some variations.
One has lost 8_267.  The other has added 7_59.
<p>
The 11 samples that do not express 39_1014 include 9 that show an history of a
recombination that has replaced 39_1014 with series from the EAS tree root including
81_857.
<p>
The other 2 samples have replaced 39_1014 and associated series with 26_1414,
64_1575, 10_2206, and 7_1868 from the EUR tree root.
</div>
