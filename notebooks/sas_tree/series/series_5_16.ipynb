<h3>Series 5_16</h3>
<div style="width:700px">
<p>
This notebook documents the series 5_16 and its 5 SNPs.  The 1000 genomes data includes
4987 samples that express none of its 5 SNPs and 16 samples that express all 5 of them.
The 5 SNP series covers more than 462,000 DNA bases.  It is over expressed in European,
American, and some South Asian populations.  It is not expressed by any samples from East
Asian or African populations.
<p>
The series 5_16 is a child of the series 8_267.  All 5_16 samples express 8_267 and its
associated hierarchy of major SAS tree root series.  The 16 samples include 15 that express
the complete SAS tree root hierarchy including 39_1014 and 1 that has experienced a recombination
event replaced 39_1014.
