<h3>Series 193_843</h3>
<div style="width:700px">
<p>
This notebook documents the series 193_843 and its 193 SNPs.  The 1000 genomes data includes
3458 samples that express none of its SNPs and 696 samples that express all 193 of them.
Another 120 samples express 192 of the SNPs. The 193 SNP series covers more than 628,000 DNA bases.
It is most over expressed in South Asian populations, modestly overexpressed in American populations,
average in East Asian populations and modestly underexpressed in African and European populations.
All 843 samples express 62_1265 and 123_1561.  842 of them express 117_685.
<p>
Another 321 samples express fragments of 193_843 more than 4 and less than 182 SNPs.  This notebook
explores the series associated with the four fragments expressed by large numbers of samples.
<p>
A 12 SNP fragment is expressed by 39 samples.  Those samples include 38 that were selected by the
series 147_38.  These samples all come from African populations.
<p>
94 samples express 29 SNP fragments. There are two different fragments.  One includes the first 29
SNPs of series 193_843 and is expressed by 93 samples.  All of these samples express the series 8_415.
Most of them express an association that includes 5_684, 32_1361, 64_1575, 10_2206, and 7_1868.  The
other sample expresses a fragment of 29 SNPs that are spread over the 193 SNP series.  It also expresses
123_1561 and 117_1685.
<p>
A 55 SNP fragment is expressed by 27 samples.  It was selected by the series 25_27 in the hierarchy
selected by 28_434 that expresses both 123_1561 and 117_1685.
<p>
A 170 SNP fragment is expressed by 41 samples.  This fragment has resulted from a recombination event
of the series 8_267 and selection by the series 9_39.
<p>
Only 16 samples express 193_843 without any of its descendant series in the SAS tree.  The hierarchies
expressed by these samples are explored in this notebook
</div>
