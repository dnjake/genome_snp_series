<h3>Series 4_815</h3>
<div style="width:700px">
<p>
This notebook documents the series 4_815 and its 4 SNPs.  The 1000 genomes data includes
4178 samples that express none of its 4 SNPs and 815 samples that express all 4 of them.
The 4 SNP series covers more than 84,000 DNA bases.  It is over expressed in South Asian 
East Asian, and American populations.
<p>
The 815 samples include 708 that express the series 6_713.  Essentially 4_815 and 6_713
form a 10 SNP series expressed by 708 samples that are part of the root of most of the
SAS tree.
<p>
Most of the other samples that express 4_815 are derived from two recombinations.  One of
them expressed by 42 samples appears to have origninated with a recombination event of
the series 8_267 that associated that series with 32_1361 and the upper region of the EAS
tree.  Those 42 samples include 39 that express the series 9_39.  The emergence of that
series appears to have selected the 4_815 recombinant for overexpression.
<p>
The other large number of samples are the 90 that express an association of 4_815 and
5_684.  84 of these samples come from East Asian populations.
<p>
The remaining 16 samples appear to show varied histories of recombination.
</div>
