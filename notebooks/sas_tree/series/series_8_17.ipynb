<h3>Series 8_17</h3>
<div style="width:700px">
<p>
This notebook documents the series 8_17 and its 8 SNPs.  The 1000 genomes data includes
4989 samples that express none of its 8 SNPs and 17 samples that express all 8 of them.
The 8 SNP series covers more than 296,000 DNA bases.  All of the 8_17 samples come from
South Asian populations.
<p>
The series 8_17 is a child of the series 29_68.  The hierarchies expressed by its samples
vary more from the standard hierarchy than most SAS tree series.  None of the samples
express 39_1014.  All 17 do express 123_1561 and 62_1265.  But 2 of the samples have
lost 29_68, 193_843, and 117_1685 through a recombination event that replaced the
lower part of the 8_17 hierarchy with the series 13_1696.
<p>
The series expressed in the upper region of the gene r3hdm1 are close to those of
the standard SAS tree root.  All 17 samples express 10_1218, 4_1149, and 13_911.
But only 11 samples express 5_1460.
<p>
All 17 samples express the series 10_2206, 9_378, and 6_99 that cover the gene ubxn4
as well as the series 15_59 that covers both the upper region of r3hdm1 and ubxn4.
<p>
The samples express a variety of less common series in the upper 120,000 DNA bases
of the studied region.
</div>
