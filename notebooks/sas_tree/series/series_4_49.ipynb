<h3>Series 4_49</h3>
<div style="width:700px">
<p>
This notebook documents the series 4_49 and its 4 SNPs.  The 1000 genomes data includes
4955 samples that express none of its 4 SNPs and 49 samples that express all 4 of them.
The 4 SNP series covers almost 88,000 DNA bases.  All 49 samples come from South Asian
populations.
<p>
The series 4_49 covers the upper part of the gene r3hdm1.  It is located beyond the end
of 193_843.  But all 49 samples express 6_713, 4_815, 193_843, 6_946, 62_1265, and 117_1685.
Those samples include 46 that express 4_416.  The samples include 33 that express all of the
common series of the SAS tree root including 39_1014.
</div>
