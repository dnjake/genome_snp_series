<p>
This notebook documents the series 9_29 and its 9 SNPs.  The 1000 genomes data includes
4974 samples that express none of its 9 SNPs and 26 samples that express all 9 of them.
Another 3 samples express 8 of the SNPs. The 9 SNP series covers more than 494,000 DNA bases.
It is overexpressed in South Asian and American populations.  It is under expressed in
African and East Asian populations.
<p>
The series 9_29 is a child 4_416.  All 29 samples express the major SAS tree hierarchy
including 4_416, 6_713, 4_815, 193_843, 6_946, 62_1265, 123_1561, and 117_1685.  24
samples express the complete SAS tree root including 39_1014.  4 samples show an history
of recombination with the upper region of EAS tree root.  1 sample shows an history of
recombination with 64_1575, 10_2206, and 7_1868.
