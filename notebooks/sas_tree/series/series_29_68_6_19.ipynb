<h3>Series 29_68_6_19</h3>
<div style="width:700px">
<p>
This notebook documents the series 6_19 that is a child of 29_68 and its 6 SNPs.  The 1000
genomes data includes 4989 samples that express none of its 6 SNPs and 19 samples that
express all 6 of them.  The 6 SNP series covers almost 314,000 DNA bases.  It is
overexpressed in South Asian populations.  It is not expressed by any samples from East
Asian or African populations.
<p>
All 19 samples express 29_68 and the four standard series that cover the lower 600,000
DNA bases of the SAS tree root.  A variety of series are expressed in the upper region
that show a substantial history of recombination.
</div>
