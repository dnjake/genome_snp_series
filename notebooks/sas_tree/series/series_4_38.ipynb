4962 samples that express none of its 4 SNPs and 38 samples that express all 4 of them.

populations.

62_1265, 123_1561, and 117_1685.  33 of the samples express the same hierarchy that
includes 4_91 and all of the common series in the complete SAS tree root.  2 samples
