<h3>Series 6_43</h3>
<div style="width:700px">
<p>
This notebook documents the series 6_43 and its 6 SNPs.  The 1000 genomes data includes
4960 samples that express none of its 6 SNPs and 37 samples that express all 6 of them.
Another 6 samples express 5 of the SNPs. The 6 SNP series covers almost 500,000 DNA bases.
It is overexpressed in South Asian and modestly overexpressed in European populations.
It is not expressed by any samples from East Asian or African populations and by only
1 sample from an American population.
<p>
The series 6_43 is a child of the series 4_416.  All 6_43 samples express 117_1685,
123_1561, 62_1265, 6_946, 193_843, 6_713, 4_815, 4_416, 5_1460, 10_1218, 4_1149,
and 13_911.  These are all the standard series of the SAS tree in the region's lower
600,000 DNA bases and in the next part of the region that includes the upper part of
the gene r3hdm1
<p>
The series expressed in the regions upper 280,000 DNA bases do show a more diverse history.
Only 4 samples express 39_1014.  22 samples express 81_857.  Most of the other samples
express series commonly associated with 81_857 but not 81_857.
</div>
