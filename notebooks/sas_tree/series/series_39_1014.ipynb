<h3>Series 39_1014</h3>
<div style="width:700px">
<p>
This notebook documents the series 39_1014 and its 39 SNPs.  The 1000 genomes data includes
3160 samples that express none of its 39 SNPs and 935 samples that express all 39 of them.
Another 72 samples express 38 of the SNPs. The 39 SNP series covers more than 190,000 DNA bases
in a similar upper part of the studied region as 64_1575 and 81_857.  It is modestly over
expressed in South Asian and American populations.  But, it is substantially expressed in
all of the 1000 genomes populations.
<p>
Six selection processes have been identified that contributed to the overexpression of
39_1014.  The most significant one was the emergence of the SAS tree.  The 843 samples
that express hierarchies that form the SAS tree include 518 that express 39_1014.
<p>
A second independent selection process was the one that generated the hierarchies associated
with the series 74_210.  The 210 samples that express 74_210 include 167 that express
39_1014.
<p>
A third independent selection process was the one that generated the series 6_88.  That
series emerged in the hiearchy that includes 28_434, 180_251, 123_1561, and 13_1696.
The 88 samples that express 6_88 include 48 that express 39_1014.
<p>
A fourth independent selection process was the one that generated the series 5_40.  That
series emerged in an hierarchy that includes 13_1696, 5_1460, 10_1218, and 4_1129. The
40 samples that express 5_40 include 27 that express 39_1014.
<p>
A fifth independent selection process was the one that generated the series 9_27.  That
series emerged in the same part of chromosome 2 as 39_1014.  All 27 samples that express
9_27 also express 39_1014.  Those samples include 16 that express the series 219_26.  There
appears to be some connection between the selection processes that generated these series
across the whole studied region.
<p>
A sixth independent selection process was the one that generated the series 4_17.  All
17 samples that express 4_17 also express 39_1014.  Those samples include 10 that express
290_16.  This process appears to have been another one that generated a complex series
that covered the whole studied region.
<p>
Another 219 samples express 39_1014.  The hiearchies expressed by these samples seem to
result from a variety of recombination events without any obvious selecting series.
<p>
Another 525 samples express fragments of 39_1014 that are more than 3 SNPs and less than
34.  There is a substantial variation to the SNPs involved in the fragments that has
not been completely sorted out.  But most of them appear to have resulted from recombination
events where either a sequence of early or late SNPs was lost.
</div>
