<h3>Series 7_1868</h3>
<div style="width:700px">
<p>
This notebook documents the series 7_1868 and its 7 SNPs.  The 1000 genomes data includes
2929 samples that express none of its 7 SNPs and 1801 samples that express all 7 of them.
Another 67 samples express 6 of them.  There are another 156 samples that express a
4 SNP fragment of 7_1868.  The 7 SNP series covers almost  94,000 DNA bases.  
<p>
The pattern of associations observed for 7_1868 is similar to that observed for 10_2206.
The 29 samples that express 64_1575 but not 7_1868 is a larger number than the 1 sample that
expresses 64_1575 but not 10_2206.  But, it is still a small number that likely reflects
the increased chance for recombination events between the end of 10_2206 and the start
of 7_1868 relative to lower parts of the studied region of chromosome 2.
<p>
The 7_1868 samples include 243 that do not express any 64_1575 SNP's.  The 163 almost
all African samples that express the series 4_163 include 161 that express 7_1868
but not 64_1575.  Those samples include 157 that do not express any 64_1575 SNP's.
Two other series 4_27 and 6_27 almost all expressed by samples from African populations
also are associated with 7_1868 but not 64_1575.  The 54 samples expressing those series
include 53 that do not express any 64_1575 SNP's.
<p>
The 156 samples that express the 4 SNP 7_1868 fragment include all 137 of the samples that
express the 17 SNP series 17_137.
</div>
