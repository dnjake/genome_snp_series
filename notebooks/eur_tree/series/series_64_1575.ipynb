<h3>Series 64_1575</h3>
<div style="width:700px">
<p>
This notebook documents the series 64_1575 and its 64 SNPs.  The 1000 genomes data includes
2689 samples that express none of its SNPs and 1369 samples that express all 64 of them.
Another 135 express 63 of them.  Another 415 samples express fragments of the 64 SNPs
that include more than 3 and less than 58 of them.  The 64 SNP series covers an almost 272,000
base region that includes the genes ubxn4, lct, mcm6, and dars. Most of the fragments appear
to have resulted from simple recombination events.
<p>
The data presented in this notebook and others shows that 64_1575, 10_2206, and 7_1818 effectively
are parts of a common series.  All but 1 of the samples that express 64_1575 also express
10_2206.  Only 29 of them fail to express 7_1868.  This notebook primarily explores the expression
of associations that include all three series but not the series 26_1414.  Most of the samples
that express those associations come from East Asian populations.  The sampled chromsomes appear to
have resulted from recombination events that replaced the series 26_1414 with the series 32_1361. 
The samples expressing the resulting association of 32_1361, 64_1575, 10_2206, and 7_1818 along
with other components of the EAS tree appear to have experienced some form of selective advantage 
that has resulted in an large overexpression of them in East Asian populations.  The selective advantage
associated with 64_1575 in a very different context from the EUR tree provides additional evidence
that its overexpression in the EUR tree is not just the chance result of linkage disequilibrium.
</div>
