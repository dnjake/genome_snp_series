<h3>Series 10_2206</h3>
<div style="width:700px">
<p>
This notebook documents the series 10_2206 and its 10 SNPs.  The 1000 genomes data includes
2511 samples that express none of its 10 SNPs and 2127 samples that express all 10 of them.
Another 79 samples express 9 of them.  A fragment of 5 of the SNPs is expressed by 194
samples.  The series covers a 76,000 DNA base region that includes the genes ubxn4 and lct.
<p>
The 2206 samples that express 10_2206 include 1574 that express 64_1575 and 1391 that
express 26_1414.  The series 26_1414, 64_1575, 10_2206, and 7_1868 all have some
history as components of a single series.  Many of the chromosomes expressing 10_2206
but not 64_1575 retain varying fragments of 64_1575 that include the SNPs from 64_1575
that overlap 10_2206.
<p>
The association of 10_2206, 64_1575, and 7_1868 forms the upper part of the EUR tree
root.  There are 1186 samples that express 10_2206 with 64_1575 and 26_1414.  Another
355 samples express 10_2206 with 64_1575 and 32_1361.  These samples appear to have
resulted from recombination events where 10_2206 and 64_1575 replaced 81_857 as the
upper region of the EAS tree.
<p>
There are 235 samples that express 10_2206 without any SNPs from 64_1575.  All 
235 of these samples also express the series 9_378.  The other 143 10_2206 9_378
associations include a variety of 64_1575 fragments.  This data suggests an history for
9_378 that started with an association of its 9 SNPs the series 10_2206, and the series
64_1575.  An history of genetic events caused varying loss of the 64_1575 SNPs with some
fragments displaying some kind of selective advantage.  The big selective advantage
came with the loss of all 64_1575 SNPs and resulted in 235 chromosomes that expressed
the association of 9_378 and 10_2206 but none of the 64_1575 SNPs.
<p>
The 194 samples that express the 5 SNP 10_2206 fragment include 160 that express the
series 6_1503, 4_1699, 4_1149, and 95_176.  The samples expressing this association
of series all come from East Asian or American populations.
</div>
