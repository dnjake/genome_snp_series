<h3>Series 4_1699</h3>
<div style="width:700px">
<p>
<p>
This notebook documents the series 4_1699 and its 4 SNPs.  The 1000 genomes data includes
3130 samples that express none of its 4 SNPs and 1699 samples that express all 4 of them
The 4 SNP series covers an almost 80,000 base region of the genes rab3gap1 and zranb3.
<p>
The 1699 samples include 1268 that express 6_1503 and 26_1414.  This association forms
the EUR tree root.  It has been selected by the several series that form EUR tree
hierarchies.
<p>
Another 231 samples express 4_1699 and 6_1503 but not 26_1414.  Those samples include
206 that express 4_1149 and 161 that express 95_176.  Some kind of selective advantage
for the recombinations that generated the association of 6_1503, 4_1699, 4_11_49,
95_176, and 51_176 seems to have been responsible for the strong overexpression of
this association in East Asian populations.
<p>
Another 200 samples express 4_1699 without 6_1503  Those samples include 169 that
resulted from five different selection events driven by the African series
18_60, 15_24, 24_24, 12_29, and 12_32.  The diversity of these events adds to the
evidence that some kind of functional value is associated with the selection
of 4_1699.
</div>
