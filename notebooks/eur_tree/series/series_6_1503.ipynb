<h3>Series 6_1503</h3>
<div style="width:700px">
<p>
This notebook documents the series 6_1503 and its 6 SNPs.  The 1000 genomes data includes
3388 samples that express none of its 6 SNPs and 1469 samples that express all 6 of them.
The 6 SNP series covers a more than 368,000 DNA base region of chromosome 2 that includes 
the genes rab3gap1 and zranb3.    
<p>
The 1000 genome samples that express 6_1503 include 51 from African populations.  But 34 of those
are the result of selection for lactase persistence that happened after the expansion of
humans from Africa.  Only 17 African samples represent some history for 6_1503
in the African past.  The series 4_1699 is in the middle of 6_1503.  Only 4 chromosome 2 samples
express 6_1503 but not 4_1699.
<p>
The association between 6_1503, 4_1699 and 26_1414 is expressed by 1268 chromosome samples.
That association defines the EUR tree root.  All 765 samples that express the series 11_765
associated with lacatse persistence express this association.
<p>
The 233 samples that express 6_1503 but not 26_1414 include 208 that express the series 4_1149.  
Those 208 samples include 163 that express the series 95_176.  All of those 163 samples come
from East Asian or American populations.
<p>
The other 25 samples include 17 that express the series 32_1361.  The plots in this notebook explore
those series.  The numbers are small.  But the data still add to the evidence of an identity for 6_1503
outside the EUR tree.
</div>
