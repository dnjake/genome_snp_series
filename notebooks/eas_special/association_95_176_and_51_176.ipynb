<h3>Associations 95_176 and 51_176</h3>
<div style="width:700px">
<p>
The association of the two series 95_176 and 51_176 spans most of the top 400,000 bases
of the studied region that includes the upper half of the gene r3hdm1 and the genes ubxn4,
lct, mcm6, and dars.   This region of chromosome 2 is also the one covered by 26_1414,
10_2206, 64_1575, and 7_1868 in the EUR tree root.  The series 95_176 and 51_176 do not
form the same kind of hierarchy as the EUR, SAS, and EAS trees.  But, the series are heavily
overexpressed in East Asian populations.
<p>
The samples that express 95_176 and 51_176 include 139 that express both of these series
in association with the series 6_1503 and 4_1699 that form the lower 368,000 bases of the
EUR tree root.  All 139 of these samples express 6_1503, 4_1699, 4_1149, 95_176, 7_1760,
51_176, 16_511, 24_1504, and 5_1296.  All 139 of them come from East Asian populations.
<p>
The association of series expressed by the 139 samples does not fit the pattern of
an hierarchy formed by some process associated with a selector series.  This association
appears to have been formed by a recombination event.  The other samples that express 95_176
and 51_176 express a variety of associations that appear to reflect a varied history of
recombination events.  The process that selected the 139 sample association for 
overexpression appears to have depended on some characteristic of the recombinant association
and not on the emergence of any new series.
<p>
The first plot shows the series most commonly associated with 96_176 and 51_176 for
the 152 samples that express both series.  Those samples include 147 from East
Asian populations and 5 from South Asian populations.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_95_176, dm.di_51_176], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```
