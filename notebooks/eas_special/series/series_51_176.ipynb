<h3>Series 51_176</h3>
<div style="width:700px">
<p>
This notebook documents the series 51_176 and its 51 SNPs. The 1000 genomes data includes 4824 
chromosome 2 samples that express none of these SNPs and 154 samples that express all 51 of them.
<p>
The 176 samples that express 51_176 include 139 that express 6_1503, 4_1699, 95_176, and 51_176.
The series 6_1503 and 4_1699 are part of the EUR tree root.  Chromosome samples that express
both 95_176 and 51_176 are analyzed in [association_95_176_and_51_176](../association_95_176_and_51_176.ipynb).
This notebook includes data on the variety of series associations for chromosomes that express
51_176 but not 95_176.  Their variety emphasizes the exceptional overexpression of the 139
samples that express 51_176 with the lower part of the EUR tree root.
</div>
