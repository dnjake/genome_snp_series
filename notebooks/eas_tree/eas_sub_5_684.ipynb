<h3>5_684 Subtree</h3>
<div style="width:700px">
<p>
The chromosome 2 samples that express this series include 444 East Asian, 145 South Asian, 
63 American, 31 European, and 1 African.  In terms of the number of samples expressing
it and the extent of its overexpression in East Asian populations, 5_684 appears to have a role
in East Asia that has some similarity with the series 11_765 that is associated with lactase
persistence in European populations.  But, the series associated with 5_684 are much more subject
to rearrangement through recombination events than those associated with 11_765.  Even when the
match criteria is relaxed to 0.8, only 13_1696, 9_944, and 32_1361 meet the criteria to be counted
as components of the 5_684 hierarchy root.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_5_684], min_match=0.8)
plt = plt_obj.do_plot()
show(plt)
```

<div style="width:700px">
<p>
This plot shows the basis plot for 5_684.  The 486 samples that express 5_684 as a basis series
include 331 East Asian, 101 South Asian, 43 American, 7 European, and 1 African.  The large number
of samples expressing this series as a basis and the high degree of population variation in its expression
make 5_684 the closest comparison to 11_765 in this interval.  But, there is a substantial difference
in the recombination behavior of the two series.  The series 11_765 associated with lactase persistence
stretches accross the whole million base studied region.  It does not appear to be associated with any
history of recombination events within that region.  In contrast a variety of recombination events have
altered the associations of 5_684 and its descendants.
</div>
