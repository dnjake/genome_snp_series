<h3>Association 13_1696 and 32_1361 Tree Root</h3>
<div style="width:700px">
<p>
The lower 700,000 base range of the EAS tree root includes the series 13_1696, 9_944,
and 32_1361.  The series 9_944 appears to be the series in the EAS tree root that is
most specific to the EAS tree.  This notebook explores the series associations expressed
by samples that express 13_1696 and 32_1361 but not 9_944 or 5_684.  
<p>
This plot shows the most common series expressed by samples that meet those criteria.
The series 4_1699 is commonly expressed by these samples instead of the series 9_944.
The 149 samples that express 4_1699 in this association include 139 African, 9 American, and
1 European.  The series 4_1699 is a standard component of the EUR tree root.  This example provides
another case of an identity for an EUR root series that transcends the EUR tree.
</div>

```python
plt_obj = dm.superset_yes_no([dm.di_13_1696, dm.di_32_1361], [dm.di_9_944, dm.di_5_684], min_match=0.5)
plt = plt_obj.do_plot()
show(plt)
```

```python
HTML(plt_obj.get_html())
```
