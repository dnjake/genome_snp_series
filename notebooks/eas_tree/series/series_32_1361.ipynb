<h3>Series 32_1361</h3>
<div style="width:700px">
<p>
This notebook documents the series 32_1361 and its 32 SNPs.  The series 32_1361 is a central
component of the EAS tree root.  It ranges over an 93,000 DNA base sequence that covers the
upper part of the gene r3hdm1.  This part of the studied region of chromosome 2 is similar
to that covered by 26_1414 in the EUR tree root.
<p>
The 1000 genomes data includes 3312 chromosome 2 samples that express none of these 32 SNPs
and 1332 samples that express all 32 of them.  Another 24 samples express 31 of them.
The fragments of the 32 SNPs include a specific 24 SNP fragment that is expressed by 134
chromosome samples.  That fragment is associated with the hierarchy that was selected by
the series 74_210.
<p>
The role of 32_1361 in the EAS tree is analyzed in the the notebooks [eas_root](../eas_root.ipynb)
and [eas_super_32_1361](../eas_super_32_1361.ipynb).  This notbook explores hierarchies
rooted in 32_1361 that are outside the EAS tree and overexpressed in African populations.
These hierarchies establish an history of events that selected the upper region of the EAS
tree before the emergence from Africa
<p>
The hierarchies selected by the series 24_24, 15_24, 8_28, and 12_29 have an
added interest.  Those hierarchies all include the series 4_1699 in their roots.
That series is a component of the EUR tree root.  It is also part of the series
expressed by 139 East Asian chromosome samples that express the series
6_1503, 4_1699, 95_176, and 51_176.
<p>
The other hierarhcy rooted in 32_1361 was selected by the series 9_39.  The root
of that hierarchy resulted from a recombination event between an hierarchy of the
SAS tree selected by the series 8_267 and the upper region of the EAS tree root.
But, no East Asian, only 1 African, and 3 South Asian samples express 9_39.  It
is overexpressed in European and American populations.
</div>
