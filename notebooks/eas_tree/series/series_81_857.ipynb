<h3>Series 81_857</h3>
<div style="width:700px">
<p>
This notebook documents the series 81_857 and its 81 SNPs. The 1000 genomes data includes
3331 samples that express none of the 81 SNPs and 583 that express all 81 of them.  Another 209
express 80 of these SNPs.  The series extends over more than 278,000 DNA bases in the same
upper part of the studied region as 64_1575 in the EUR tree.  Like 64_1575, a relatively large
number of samples do express some fragment of 81_857.
<p>
All 857 81_857 samples express the series 24_1504, 4_1442, 7_1303, and 5_12
96. 856 of the samples
express 7_1760 and 6_1114.  850 express 13_1227.  This pattern is consistent with the idea that
all of these series have an history as a single series of SNPs that has fragmented over time.
<p>
The 857 samples include 781 that express 32_1361, 546 that express 9_944, and 259 express 5_684.
<p>
The basis plot for 81_857 without 32_1361, 9_944, or 5_684  does not show much sign of any specific
selective advantage.  The 76 samples that express 81_857 without any of these series appear
to have resulted from a variety of recombination events.
<p>
The basis plot for 81_857 and 32_1361 without 9_944 or 5_684 shows an history of several selection
events.  Several of them involve hierarchies selected by series expressed by samples from African
populations.
<p>
The basis plot for 81_857, 32_1361, 9_944 without 5_684 shows an history of several additional selection
events.  The 150 samples that only express 81_857, 32_1361, and 9_944 are one example.  Significant numbers
of samples from African and East Asian populations are part of this example.  But, both of those populations
are underexpressed.  The biggest overexpression is for the 38 samples from American populations.  European
and South Asian populations are also overexpressed.  Samples from African and East Asian populations are
absent from most of the other selective events from this plot.
<p>
The last basis plot is for 81_857 and 5_684.  The 79 samples from this plot that express either 4_67 or 5_47
as a basis are very strongly overexpressed for East Asian populations.  The 87 samples from this plot that
express 5_684 for a basis are overexpressed for East Asian, American, and South Asian populations.  Only 1
of the samples comes from Europe and none from Africa.  This plot understates the role of 5_684 and its
East Asian overexpression because many of the East Asian samples that express 5_684 as a basis are derived
from recombination events that replaced 81_857 with 64_1575, 10_2206, and 7_1868.  The 22 samples from this
plot that express the series 4_25 as a basis all come from East Asian populations.
<p>
The population distributions are very different for the 93 samples from this plot that express the
series 7_59, 14_43, 4_37, or 4_35 as a basis.  None of those samples come from East Asian populations.
</div>
