<h3>Series 5_47</h3>
<div style="width:700px">
<p>
This notebook documents the series 5_47 and its 5 SNPs. The 1000 genomes data sample
SNP counts for 5_17 are 4952 samples 0 SNPs, 1 samples 1 SNP, 2 samples 2 SNPs, 2 samples 3
SNPs, 4 samples 4 SNPs, and 47 samples 5 SNPs.  The 5 SNP series extends over more than
347,000 DNA bases.  
<p>
The series 5_47 is a child of 5_684.  All but three of the 47 samples express 5_684.
All of those three have lost the top 5_684 SNP.
<p>
The samples that express 5_47 are very overexpressed for East Asian populations.
The samples include 42 from East Asia and 5 from South Asia.  The series is
not expressed by any African, American, or European samples.
<p>
38 of the samples express 81_857 in the upper region of the studied interval.  Only
1 expresses 64_1575.  These numbers distinguish the subset of samples that express
5_684 and 5_47 from those that only express 5_684.
<p>
16 of the samples do express 4_815 in the lowest part of the region instead of
16_1396.  This behavior is similar to samples that only express 5_684.
</div>
